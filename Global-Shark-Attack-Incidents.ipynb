```python
null_cols = df.isnull().sum()
print(null_cols[null_cols > 0])
df.shape
```

```python
'''
LIMPIANDO COLUMNAS
Elegimos las columnas con datos relevantes a mi estudio. 
Los pdf, columnas repetidas y los nombres de las personas afectadas no me interesan,por ello las eliminamos.
Además eliminamos las columnas que tienen mas de 4500 datos nulos, visto que en relacion a 
los 5992 datos que tenemos no nos aportan ninguna información.
'''
drop_cols = list(null_cols[null_cols > 4500].index)
data_f = df.drop(drop_cols, axis=1)
```

```python
drop_cols = ['Name', 'pdf', 'href formula', 'href',
       'Case Number.1', 'Case Number.2']
data_f = data_f.drop(drop_cols, axis=1)
print(data_f.shape)
data_f.head(10)
```
