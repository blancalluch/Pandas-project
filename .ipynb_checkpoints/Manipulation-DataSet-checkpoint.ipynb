```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv('./Output/SharckAttack_dfClean.csv') 
```

```python
'''Estudiamos si en los últimos años han sobrevivido más  personas a los ataques de tiburón.'''
años=df['Year'].value_counts().sort_index()

cutoffs = list(range(1900,2035,15))
year_labels = ['1900-15', '1915-30', '1930-45', '1945-60', '1960-75','1975-90','1990-05','2005-20']
bins = pd.cut("Year",cutoffs, labels=year_labels)
df["Year"] = bins
#cutoffs
bins
```

```python
df2=df[['Year','Country','Activity','Sex ','Age','Fatal (Y/N)']]
d_C=df2.groupby(['Country']).count()
d_Y=df2.groupby(['Year']).count()
d_A=df2.groupby(['Activity']).count()
print(df2.shape)
display(d_C)
display(d_A)
```
